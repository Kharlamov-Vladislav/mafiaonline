```python
import requests
import json
from pprint import pprint
def get_log(id_game):
    """–í–æ–∑—Ä–∞—â–∞–µ—Ç –ø–æ–ª–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –ø–∞—Ä—Ç–∏–∏:
    ----------------------------------------------------
    –í–•–û–î–ù–´–ï –ü–ê–†–ê–ú–ï–¢–†–´:
    id_game :: –¶–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É–Ω–∏–∫–∞–ª—å–Ω–æ–≥–æ –∏–Ω–¥–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ –ø–∞—Ä—Ç–∏–∏
    
    –í–û–ó–†–ê–©–ê–ï–¢:
    –ü–æ–ª–Ω—ã–π –ª–æ–≥ –ø–∞—Ä—Ç–∏–∏ –≤ —Ñ–æ—Ä–º–∞—Ç–µ —Å–ª–æ–≤–∞—Ä—è
    """
    
    url = f'https://www.mafiaonline.ru/api/api.php?action=log&param=log&id={id_game}'
    r = requests.get(url)
    status_code = r.status_code
    r = json.loads(r.text)
    if status_code == 200 and r['r'] == 'ok':
        return r['log']
```

```python
chat = get_log(4226783)['chat'] # –ß–∞—Ç –ø–∞—Ä—Ç–∏–∏
summary = get_log(4226783)['summary'] # –•–æ–¥—ã –∏–≥—Ä–æ–∫–æ–≤
log = get_log(4226783) # –ü–æ–ª–Ω–æ—Å—Ç—å—é –ª–æ–≥ –ø–∞—Ä—Ç–∏–∏
```
